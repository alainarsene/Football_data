```python
import pandas as pd
```

```python
df1=pd.read_csv('PremierleagueValeur.csv')
df1
```

```python
df2=pd.read_csv('premier_league.csv')
df2
```

```python
df2=df2.drop(['_id','team.crest','team.id','team.name','team.tla'], axis=1)
df2
```

```python
df2=df2.rename(columns={'team.shortName':'Club'})
df2
```

```python
Angleterre = pd.merge(df1, df2, on='Club', how='right')
Angleterre
```

```python
Angleterre.to_csv('Angleterre.csv', index=False)
```

```python
#Pearson
correlation = Angleterre['won'].corr(Angleterre['Valeurs'])
print(correlation)
```

```python
#Kendall
correlation = Angleterre['won'].corr(Angleterre['Valeurs'], method='kendall')
print(correlation)
```

```python
#Spearman
correlation = Angleterre['won'].corr(Angleterre['Valeurs'], method='spearman')
print(correlation)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
h_labels = [x.replace('_', ' ').title() 

for x in list(France.select_dtypes(include=['number', 'bool']).columns.values)]

fig, ax = plt.subplots(figsize=(10,6))
_ = sns.heatmap(
France.corr(), 
annot=True, 
xticklabels=h_labels, 
yticklabels=h_labels, 
cmap=sns.cubehelix_palette(as_cmap=True), ax=ax)
#Forte correlation entre la variable Valeurs et Goalsfor(0.75), Points(0.66), Won(0.69)
```

```python
#Pearson correlation: La corrélation de Pearson est utilisée pour mesurer la corrélation linéaire entre deux variables continues
France.corr(method ='pearson')
```

```python
#Kendall correlation: La corrélation de Kendall est utilisée pour mesurer la corrélation ordinale entre deux variables
France.corr(method ='kendall')
```

```python
#Spearman correlation: La corrélation de Spearman est utilisée pour mesurer la corrélation monotone entre deux variables
France.corr(method ='spearman')
```

```python
#Calculer la correlation entre deux variables et afficher
subset_df = France[['won', 'Valeurs']]

# Calculer la corrélation entre les variables
corr = subset_df.corr()

# Afficher le graphique de corrélation sous forme de heatmap
sns.heatmap(corr, annot=True)
plt.show()
```
